```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import make_scorer, recall_score
```

<div style="background:#9eb5db;border-radius:15px;"><h2 style="line-height:50px;height:50px;margin-left:20px;color:white">Problem
</h2></div>

```python
df = pd.read_csv("heart.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# We Don't Have Null or Nan Colums But We Want to oneHotEncode Slope Col and Scaling all cols
```

```python
X = df.drop('target', axis=1)
y = df['target']
```

```python
ct = ColumnTransformer([('slope', OneHotEncoder(), ['slope'])], remainder='passthrough')
X_encoded = ct.fit_transform(X)
X_encoded
```

```python
# Scale Dataset

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split Dataset

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state=222)
```

```python
#  { Decision Tree }

scorer = make_scorer(recall_score, pos_label=1)
params = {
    'criterion': ['gini', 'entropy'],
}

## 1: We are Looking for the most f1

gsh = GridSearchCV(DecisionTreeClassifier(), param_grid=params, scoring='f1', cv=3, n_jobs=-1, verbose=10)
gsh.fit(X_train, y_train)

print(gsh.best_estimator_)
print(gsh.best_score_)
print(gsh.best_params_)

clf = DecisionTreeClassifier(criterion=gsh.best_params_['criterion'])
clf.fit(X, y)
y_pred = clf.predict(X_test)
print(metrics.classification_report(y_test, y_pred))

## 2: We are Looking for the most recall for class 1

gsh = GridSearchCV(DecisionTreeClassifier(), param_grid=params, scoring=scorer, cv=3, n_jobs=-1, verbose=10)
gsh.fit(X_train, y_train)

print(gsh.best_estimator_)
print(gsh.best_score_)
print(gsh.best_params_)

clf = DecisionTreeClassifier(criterion=gsh.best_params_['criterion'])
clf.fit(X, y)
y_pred = clf.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
```

```python
# { Random Forest }

scorer = make_scorer(recall_score, pos_label=1)
params = {
    'n_estimators': [500, 1000, 1500, 2000],
    'bootstrap': [True, False],
    'oob_score': [True, False]
}

## 1: We are Looking for the most f1

gsh = GridSearchCV(RandomForestClassifier(), param_grid=params, scoring='f1', cv=3, n_jobs=-1, verbose=10)
gsh.fit(X_train, y_train)

print(gsh.best_estimator_)
print(gsh.best_score_)
print(gsh.best_params_)

rfc_clf = RandomForestClassifier(n_estimators=gsh.best_params_['n_estimators'],
                                 bootstrap=gsh.best_params_['bootstrap'], oob_score=gsh.best_params_['oob_score'])
rfc_clf.fit(X, y)
y_pred = rfc_clf.predict(X_test)
print(metrics.classification_report(y_test, y_pred))

## 2: We are Looking for the most recall for class 1

gsh = GridSearchCV(RandomForestClassifier(), param_grid=params, scoring=scorer, cv=3, n_jobs=-1, verbose=10)
gsh.fit(X_train, y_train)

print(gsh.best_estimator_)
print(gsh.best_score_)
print(gsh.best_params_)

rfc_clf = RandomForestClassifier(n_estimators=gsh.best_params_['n_estimators'],
                                 bootstrap=gsh.best_params_['bootstrap'], oob_score=gsh.best_params_['oob_score'])
rfc_clf.fit(X, y)
y_pred = rfc_clf.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
```

```python
# { AdaBoost }

score = make_scorer(recall_score, pos_label=1)
params = {
    'n_estimators':[400, 800, 1000, 1500],
    'base_estimator': [DecisionTreeClassifier(),
                       KNeighborsClassifier(n_neighbors=7),
                       SVC(probability=True, kernel='rbf'),
                       RandomForestClassifier(n_estimators=1000, bootstrap=True, oob_score=True)]
}

## 1: We are Looking for the most f1

gsh = GridSearchCV(AdaBoostClassifier(), param_grid=params, scoring='f1', cv=3, n_jobs=-1, verbose=10)
gsh.fit(X_train, y_train)

print(gsh.best_estimator_)
print(gsh.best_score_)
print(gsh.best_params_)

clf = AdaBoostClassifier(n_estimators=gsh.best_params_['n_estimators'], base_estimator=gsh.best_params_['base_estimator'])
clf.fit(X, y)
y_pred = rfc_clf.predict(X_test)
print(metrics.classification_report(y_test, y_pred))


## 2: We are Looking for the most recall for class 1


gsh = GridSearchCV(AdaBoostClassifier(), param_grid=params, scoring=score, cv=3, n_jobs=-1, verbose=10)
gsh.fit(X_train, y_train)

print(gsh.best_estimator_)
print(gsh.best_score_)
print(gsh.best_params_)

clf = AdaBoostClassifier(n_estimators=gsh.best_params_['n_estimators'], base_estimator=gsh.best_params_['base_estimator'])
clf.fit(X, y)
y_pred = rfc_clf.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
```

```python
# { KNN }

score = make_scorer(recall_score, pos_label=1)
params = {
    'n_neighbors': [3, 5, 7, 10],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
}

## 1: We are Looking for the most f1

gsh = GridSearchCV(KNeighborsClassifier(), param_grid=params, scoring='f1', cv=3, n_jobs=-1, verbose=10)
gsh.fit(X_train, y_train)

print(gsh.best_estimator_)
print(gsh.best_score_)
print(gsh.best_params_)

clf = KNeighborsClassifier(n_neighbors=gsh.best_params_['n_neighbors'], algorithm=gsh.best_params_['algorithm'])
clf.fit(X, y)
y_pred = clf.predict(X_test)
print(metrics.classification_report(y_test, y_pred))

## 2: We are Looking for the most recall for class 1

gsh = GridSearchCV(KNeighborsClassifier(), param_grid=params, scoring=score, cv=3, n_jobs=-1, verbose=10)
gsh.fit(X_train, y_train)

print(gsh.best_estimator_)
print(gsh.best_score_)
print(gsh.best_params_)

clf = KNeighborsClassifier(n_neighbors=gsh.best_params_['n_neighbors'], algorithm=gsh.best_params_['algorithm'])
clf.fit(X, y)
y_pred = clf.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
```

```python
# { SVM }

score = make_scorer(recall_score, pos_label=1)
params = {
    'C': [1, 5, 10, 15],
    'kernel': ['linear'],
    'gamma': ['auto', 0.1, 0.5]}

## 1: We are Looking for the most f1

gsh = GridSearchCV(SVC(), param_grid=params, scoring='f1', cv=3, n_jobs=-1, verbose=10)
gsh.fit(X_train, y_train)

print(gsh.best_estimator_)
print(gsh.best_score_)
print(gsh.best_params_)

clf = SVC(C=gsh.best_params_['C'], kernel=gsh.best_params_['kernel'], gamma=gsh.best_params_['gamma'], random_state=3244)
clf.fit(X, y)
y_pred = clf.predict(X_test)
print(metrics.classification_report(y_test, y_pred))



## 2: We are Looking for the most recall for class 1

gsh = GridSearchCV(SVC(), param_grid=params, scoring=score, cv=3, n_jobs=-1, verbose=10)
gsh.fit(X_train, y_train)

print(gsh.best_estimator_)
print(gsh.best_score_)
print(gsh.best_params_)

clf = SVC(C=gsh.best_params_['C'], kernel=gsh.best_params_['kernel'], gamma=gsh.best_params_['gamma'], random_state=3244)
clf.fit(X, y)
y_pred = clf.predict(X_test)
print(metrics.classification_report(y_test, y_pred))
```
